# Predicción continua

###### Descripción de los datos

En total hay 21 columnas el los datos, las cuáles se describen a continuación:

  - ID: Correlativo númerico que distingue cada una de las casas vendidas.
  - Date: Fecha en la cual fue vendida la casa.
  - Price: Precio de venta de la casa (en dólares norteamericanos).
  - Bedrooms: Número de cuartos en la casa.
  - Bathrooms: Número de baños disponibles en la casa, donde un valor de 0.5 representa un baño con inodoro pero si
  - Sqft_living15: Los pies cuadrados de espacio habitable de la casa interior para los 15 vecinos más cercanos.
  - Sqft_lot15: Los metros cuadrados de los terrenos de los 15 vecinos más cercanos.n ducha.
  - Sqft_living: Número de pies cuadrados del espacio habitable de la casa.
  - Sqft_lot: Pies cuadrados del espacio total del terreno donde se ubica la casa.
  - Floors: Cantidad de pisos en la casa.
  - Waterfront: Variable que indica la presencia o no de vista al mar en la casa.
  - View: Índice del 0 al 4 que indica que tan buena es la vista de la propiedad.
  - Condition: Índice del 1 al 5 para calificar la condición actual de la casa.
  - Grade: Índice del 1 al 13, el cuál califica el nivel de calidad de construcción de la casa.
  - Sqft_above: Cantidad de pies cuadrados del espacio interior de la casa. que está sobre el nivel del suelo.
  - Sqft_basement: Los pies cuadrados del espacio interior de la casa. que está por debajo del nivel del suelo.
  - Yr_built: El año en fue construida la casa.
  - Yr_renovated: Año de la última renovación de la casa.
  - Zipcode: Código postal del areá donde se encuentra la casa.
  - Lat: Latitud de la ubicación de la casa.
  - Long: Longitud de la15 vecinos más cercanos.

## Regresión Lineal Múltiple

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.feature_selection import RFE # función de selección backward
```

### Dividir datos en testing y training

1. Utilizando scikit-learn, dividir los datos en un 75% training y el resto testing. Partir del dataframe con las columnas detectadas como relevantes en el análisis exploratorio

```python
# Convertir la columna 'date' a formato datetime
data["date"] = pd.to_datetime(data["date"], format="%Y%m%dT%H%M%S")
data.head()
```

```python
correlation_matrix = data.corr().round(2)
filtro = correlation_matrix.price.abs() > 0.2
data = data.loc[:,filtro]
data.head()
```

```python
# Seleccionar variables independientes (predictoras)
#X = data[['sqft_living', 'grade', 'sqft_above']]
X = data.loc[:, ~data.columns.isin(['price'])]
# Seleccionar la variable dependiente (objetivo)
y = data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, train_size=0.75)
```

### Análisis de Multicolinealidad

2. Calcular el determinante de la matriz de correlación y determinar si hay multicolinealidad

```python
matriz = data.corr().round(2)
matriz
```

```python
# Calcular el determinante de la matriz de correlación
data_filtrada = data.loc[:, data.columns != 'price']
matriz = data_filtrada.corr()
determinante = np.linalg.det(matriz)
print(f"\nDeterminante de la matriz de correlación: {determinante:}")
# DET = 0 -> hay correlación perfecta entre las variables -> CORRELACIÓN CERCANA A 1
#Si el determinante ≈ 0 → Hay multicolinealidad grave, lo que puede causar problemas en modelos de regresión (hay que quitar columnas)
#Si el determinante > 0.01 → No hay multicolinealidad fuerte.
```

3. Detectar variables a descartar con el p-valor (no eliminarlas)

Si e valor de p-valor e smenos a 0.5 --> refuto que beta sea 0 --> la X asocuada a ese Beta es significatiba

```python
X = sm.add_constant(X)  # Agregar intercepto

# Ajustar el modelo de regresión lineal
modelo = sm.OLS(y, X).fit()

# Mostrar resultados
modelo.pvalues < 0.05
```

4. Analizar VIF de variables relevantes

```python
# VIF (Factor de Inflación de la Varianza)

# SI VIF = 1 -> no hay relación entre todas las columnas y la columna analizada
# SI VIF >1 y <5 -> la relación es muy baja
# SI VIF >5 -> la relación es alta y hay que eliminar esta columna
# SI VIF = infinito -> la relación es perfecta y hay que eliminar esta columna
```

Que columnas se pueds explciar a traves de otra?

```python
vifs = {}
for i in X_train.columns:
    X_vif = X_train.loc[:, X_train.columns != i]
    y_vif = X_train.loc[:, i]
    
    lm = LinearRegression()
    lm.fit(X_vif, y_vif)
    r2 = lm.score(X_vif, y_vif)
    
    VIF = 1 / (1-r2) if r2 < 1 else np.inf
    vifs[i] = VIF

vifs
```

5. Eliminar variables con VIF mayor a 5 y recalcular. Eliminar variable de a una y verificar si se corrije el VIF del resto antes de eliminar otra

```python
# columnas con VIF mayor a 5
filtered_vifs = {k: v for k, v in vifs.items() if v > 5}
filtered_vifs
```

```python
# Quitamos 'sqft_living' y recalculamos
X = data.loc[:, ~data.columns.isin(['price', 'sqft_living'])]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, train_size=0.75)
```

```python
vifs = {}
for i in X_train.columns:
    X_vif = X_train.loc[:, X_train.columns != i]
    y_vif = X_train.loc[:, i]
    
    lm = LinearRegression()
    lm.fit(X_vif, y_vif)
    r2 = lm.score(X_vif, y_vif)
    
    VIF = 1 / (1-r2) if r2 < 1 else np.inf
    vifs[i] = VIF

vifs
```

```python
filtered_vifs = {k: v for k, v in vifs.items() if v > 5}
filtered_vifs
```

6. Seleccionar variables a utilizar en el modelo con LinearRegression y RFE (Probar con el número maximo de columnas menos 1 y el numero maximo de columnas). Luego calcular el R^2 del modelo ajustado con los datos de training, y evaluarlo en los de testing

```python
# va a evitarme calcular todas las combinaciones posibles de regresiones lineales
selector = RFE( # función que selecciona el mejor modelo (Recursive Feature Elimination) 
    LinearRegression(), # Que modelo? Una regresion lineal
    n_features_to_select = 10 # cuantas variables querés seleccionar? cuantas columnas: 10
)
selector = selector.fit(X_train, y_train)
columnas_a_seleccionar = selector.support_
columnas_a_seleccionar
```

```python
X_train_final = X_train.loc[:, columnas_a_seleccionar]
X_test_final = X_test.loc[:, columnas_a_seleccionar]
```

```python
# Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(X_train_final, y_train)

y_pred = modelo.predict(X_test_final)
y_pred
```

```python
# Evaluación del modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f'Coeficientes: {modelo.coef_}')
print(f'Intercepto: {modelo.intercept_}')
print(f'Error Cuadrático Medio (MSE): {mse:.4f}')
print(f'Coeficiente de Determinación (R²): {r2:.4f}')
```
