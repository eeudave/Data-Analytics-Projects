```python
from scipy.spatial import distance_matrix
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
from sklearn.preprocessing import MinMaxScaler
from scipy.cluster.hierarchy import fcluster
```

# Clustering Jerarquico

En este ejemplo analizaremos los datos de ventas de una empresa. En la tabla tendremos información de clientes, cada fila consta con la información de clicks y compras realizados por cada uno

```python
np.random.seed(1) # genera que siempre den igual los resultados
clicks = np.concatenate(
    (np.random.uniform(1,5,10),
     np.random.uniform(7,10,10),
     np.random.normal(3,1,10),
     np.random.normal(8.5,1,10))
).round(2)
compras = np.concatenate(
    (np.random.uniform(1,5,10),
     np.random.normal(3,1,10),
     np.random.uniform(7,10,10),
     np.random.normal(8.5,1,10))
).round(0)

data = pd.DataFrame(
    {
        "Clicks": (clicks*100).astype("int"),
        "Compras": compras.astype("int")
    }
)
data.head()
```

1. Graficar datos en scatterplot

```python
plt.figure(figsize=(4,4
                   ))
plt.grid(True, linestyle="--", alpha=0.6)
sns.scatterplot(
    x = data["Clicks"], 
    y = data["Compras"], 
    s = 200
)
plt.savefig("1.png")  # Guarda en formato PNG
plt.show()
```

2. Normalizar la data

```python
# Normalización Min-Max
scaler = MinMaxScaler()
data_normalized = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
data_normalized
```

3. Graficar nuevamente los datos, ¿nota algún cambio?

```python
plt.figure(figsize=(4,4))
plt.grid(True, linestyle="--", alpha=0.6)
sns.scatterplot(
    x = data_normalized["Clicks"], 
    y = data_normalized["Compras"], 
    s = 200
)
plt.show()
```

4. Graficar dendograma con el método del centroide utilizando la distancia euclidea

```python
Z = linkage(
    data_normalized, 
    method= "centroid", # busca la distancia mínima o mas cercana
    metric = "euclidean" # usa la distancia euclidea
)
Z
```

```python
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("Compras por click")
plt.ylabel("Distancia")
dendrogram(Z)
plt.savefig("2.png") 
plt.show()
```

5. Calcular coeficiente de Cophonet

```python
distancia_original = pdist(data[["Clicks", "Compras"]])

c, coph_dist = cophenet(Z, distancia_original)
c
```

Al tener un valor cercano a 1, definimos que los datos son aptos para ser clusterizados.

6. Determinar número de clusters con el método del codo

```python
last = Z[:,2] # me interesa únicamente la columna de distancias (la 2)
last_rev = last[::-1]
last_rev = last_rev[:10] #10 distancias mas grandes
print(last_rev)
```

```python
plt.plot(
    range(1, len(last_rev)+1),
    last_rev
)
plt.savefig("3.png") 
```

Podemos observar que el "codo" se genera en 4 clusters, por lo que será la cantidad a seleccionar.

7. Generar 3 clusters y mostrar el resultado en un dendograma con la línea horizontal mostrando la línea de corte. También mostrar un scatterplot con los clusters

```python
plt.title("Dendrograma jerárquico para el Clustering")
plt.xlabel("Compras por click")
plt.ylabel("Distancia")
dendrogram(Z)

plt.hlines(0.6,0,1000, color = "black", ls = "--")
plt.savefig("4.png") 
plt.show()
```

```python
max_d=0.6
clusters = fcluster(Z, max_d, criterion="distance")
clusters
```

```python
data_normalized["cluster"] = clusters
data_normalized.head()
```

```python
plt.figure(figsize=(4,4))
plt.grid(True, linestyle="--", alpha=0.6)
sns.scatterplot(
    x = data_normalized["Clicks"], 
    y = data_normalized["Compras"], 
    s = 200,
    hue = data_normalized["cluster"],
    palette = "tab10"
)
plt.savefig("5.png") 
plt.show()
```
