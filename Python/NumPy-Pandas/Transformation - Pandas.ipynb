1. Joinear las 3 tablas en un único dataframe llamado "data".



```python
country_sport = data_country.merge(
    data_sports,
    left_on = ['Athlete'],
    right_on = ['Athlete'],
    how = 'left'
)

data = country_sport.merge(
    medals,
    left_on = ['Athlete'],
    right_on = ['Athlete'],
    how = 'left'
)
```

```python
data
```

3. Mostrar la suma de todas las columnas de medallas por año.

```python
summedallas = data.groupby('Year')[['Gold Medals','Silver Medals','Bronze Medals','Total Medals']].sum().reset_index()
summedallas
```

4. Mostrar el promedio de todas las columnas de medallas por año.

```python
data_agrupada = (
    data.
        groupby(["Athlete", "Age"]).
        agg(TotalMedals = ("Total Medals", "sum")).
        sort_values(["Athlete", "Age"]).
        reset_index()
)    
# 

data_agrupada["Acumulado"] = (
    data_agrupada.
        groupby("Athlete").
        agg(Acumulado = ("TotalMedals", "cumsum"))
)
# data_agrupada.loc[data_agrupada.Athlete == "Michael Phelps", :]

data_agrupada.loc[data_agrupada["Acumulado"] >= 10, :].sort_values("Age")
```

22. Hay algún deportista que en su nombre contenga una palabra palindroma (se lee igual que atrás hacia adelante)? Cual/Cuales?

```python
# Crear DataFrame con palabras
df = pd.DataFrame({
    "Palabra": ["radar", "python", "reconocer", "casa", "oso", "sol"]
})

# Función para verificar si una palabra es palíndroma
def es_palindromo(palabra):
    return palabra.lower() == palabra.lower()[::-1]  # Compara palabra con su inversa

# Filtrar solo las palabras palíndromas
df_palindromos = df[df["Palabra"].apply(es_palindromo)]

print(df_palindromos)
```
