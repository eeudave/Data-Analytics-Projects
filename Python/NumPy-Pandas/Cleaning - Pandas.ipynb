6. A partir de una lista de tuplas generada con la función zip

````python

Fruta = ["Manzana", "Naranja", "Banana", "Limon"]
Color = ["rojo", "naranja", "amarillo", "amarillo"]
Precio = [1.5, 2, 3.4, 2.6]
Costo = [1,1.5,3,2]

lfruta = list(zip(Fruta, Color, Precio, Costo))
lfruta
````

# 14. Contar cuantos registros no nulos hay en cada fila y almacenarlo en un diccionario que tenga de clave el indice de la fila

17. Realiza una copia de la data en el objeto "data_bis" para no modificar la data original y completa los valores faltantes de la columna "jornadas" con el valor más frecuente de dicha columna
